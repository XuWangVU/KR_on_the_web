```python
import csv
import glob
print(glob.glob("*.csv"))
import logging
logging.basicConfig()

from rdflib import Graph,Namespace,URIRef, Literal
from rdflib.namespace import RDF, FOAF, OWL, RDFS, DCTERMS, XSD

# amco=Namespace("http://www.gemeentegeschiedenis.nl/amco/")
# gemeentanaam=Namespace("http://www.gemeentegeschiedenis.nl/gemeentenaam/")
# cbs=Namespace("http://www.gemeentegeschiedenis.nl/cbscode/")
n = Namespace("http://gemeenteraad.org/vocabulary/")
np=Namespace("http://gemeenteraad.org/party/")
dbp=Namespace("http://dbpedia.org/resource/")
dbo =Namespace("http://dbpedia.org/ontology/")
g = Graph()
g.add((n.CBScode,RDF.type, RDF.Property))
g.add((n.CBScode,DCTERMS.description,Literal("CBS code")))

#KVP
g.add((np.KVP,RDF.type,dbo.party))
g.add((np.KVP,DCTERMS.description,Literal("Catholic People's Party")))
g.add((np.KVP,OWL.sameAs,dbp["Catholic_People's_Party"]))
#VVD
g.add((np.VVD,RDF.type,dbo.party))
g.add((np.VVD,DCTERMS.description,Literal("Liberals (VVD)")))
g.add((np.VVD,OWL.sameAs,dbp['People%27s_Party_for_Freedom_and_Democracy']))
#CPN
g.add((np.CPN,RDF.type,dbo.party))
g.add((np.CPN,DCTERMS.description,Literal("Liberals (VVD)")))
g.add((np.CPN,OWL.sameAs,dbp['Communist_Party_of_the_Netherlands']))
#D66
g.add((np.D66,RDF.type,dbo.party))
g.add((np.D66,DCTERMS.description,Literal("Democrats (D66)")))
g.add((np.D66,OWL.sameAs,dbp['Democrats_66']))
g.add((np.D66,OWL.sameAs,n["Democraten_66_(D66)"]))
g.add((np["Democraten_66_(D66)"],OWL.sameAs,n.D66))
g.add((np["Democraten_66_(D66)"],RDF.type,dbo.party))
g.add((np["Democraten_66_(D66)"],DCTERMS.description,Literal("Democrats (D66)")))
g.add((np["Democraten_66_(D66)"],OWL.sameAs,dbp['Democrats_66']))
#PvdA
g.add((np.PvdA,RDF.type,dbo.party))
g.add((np.PvdA,DCTERMS.description,Literal("Labor party (PvdA)")))
g.add((np.PvdA,OWL.sameAs,dbp['Labour_Party_(Netherlands)']))
#ARP
g.add((np.ARP,RDF.type,dbo.party))
g.add((np.ARP,DCTERMS.description,Literal("Anti-Revolutionary Party")))
g.add((np.ARP,OWL.sameAs,dbp['Anti-Revolutionary_Party']))
#CHU
g.add((np.CHU,RDF.type,dbo.party))
g.add((np.CHU,DCTERMS.description,Literal("Christelijk-Historische Unie")))
g.add((np.CHU,OWL.sameAs,dbp['Christian_Historical_Union']))
#PvdV
g.add((np.PvdV,RDF.type,dbo.party))
g.add((np.PvdV,DCTERMS.description,Literal("Freedom Party (Netherlands)")))
g.add((np.PvdV,OWL.sameAs,dbp['Freedom_Party_(Netherlands)']))
#SGP
g.add((np.SGP,RDF.type,dbo.party))
g.add((np.SGP,DCTERMS.description,Literal("Reformed Political Party")))
g.add((np.SGP,OWL.sameAs,dbp['Reformed_Political_Party']))
#RK
g.add((np.RK,RDF.type,dbo.party))
#PCG
g.add((np.PCG,RDF.type,dbo.party))
g.add((np.PCG,DCTERMS.description,Literal("Protestant-Christelijke Groepering")))
#KNP
g.add((np.KNP,RDF.type,dbo.party))
g.add((np.KNP,DCTERMS.description,Literal("Catholic National Party")))
g.add((np.KNP,OWL.sameAs,dbp['Catholic_National_Party']))
#GG
g.add((np.GG,RDF.type,dbo.party))
#GPV
g.add((np.GPV,RDF.type,dbo.party))
g.add((np.GPV,DCTERMS.description,Literal("Reformed Political League")))
g.add((np.GPV,OWL.sameAs,dbp['Reformed_Political_League']))
#PC
g.add((np.PC,RDF.type,dbo.party))
#PSP
g.add((np.PSP,RDF.type,dbo.party))
g.add((np.PSP,DCTERMS.description,Literal("Pacifist Socialist Party")))
g.add((np.PSP,OWL.sameAs,dbp['Pacifist_Socialist_Party']))
#BP
g.add((np.BP,RDF.type,dbo.party))
g.add((np.BP,DCTERMS.description,Literal("Farmers' Party (Netherlands)")))
g.add((np.BP,OWL.sameAs,dbp['Farmers%27_Party_(Netherlands)']))
#CDA
g.add((np.CDA,RDF.type,dbo.party))
g.add((np.CDA,DCTERMS.description,Literal("Christian Democratic Appeal")))
g.add((np.CDA,OWL.sameAs,dbp['Christian_Democratic_Appeal']))
#PAK
g.add((np.PAK,RDF.type,dbo.party))
g.add((np.PAK,DCTERMS.description,Literal("Progressieve samenwerking in Nederland in de jaren 1970")))
g.add((np.PAK,OWL.sameAs,dbp['Progressieve_samenwerking_in_Nederland_in_de_jaren_1970']))
#PPR
g.add((np.PPR,RDF.type,dbo.party))
g.add((np.PPR,DCTERMS.description,Literal("Political Party of Radicals")))
g.add((np.PPR,OWL.sameAs,dbp['Political_Party_of_Radicals']))
#DS70
g.add((np.DS70,RDF.type,dbo.party))
g.add((np.DS70,DCTERMS.description,Literal("Democratic Socialists '70")))
g.add((np.DS70,OWL.sameAs,dbp['Democratic_Socialists_%2770']))
#RPF
g.add((np.RPF,RDF.type,dbo.party))
g.add((np.RPF,DCTERMS.description,Literal("Reformatory Political Federation")))
g.add((np.RPF,OWL.sameAs,dbp['Reformatory_Political_Federation']))
#CD
g.add((np.CD,RDF.type,dbo.party))
g.add((np.CD,DCTERMS.description,Literal("Centre Democrats (Netherlands)")))
g.add((np.CD,OWL.sameAs,dbp['Centre_Democrats_(Netherlands)']))
#SP
g.add((np.SP,RDF.type,dbo.party))
g.add((np.SP,DCTERMS.description,Literal("Socialist Party (Netherlands)")))
g.add((np.SP,OWL.sameAs,dbp['Socialist_Party_(Netherlands)']))
g.add((np.SP,OWL.sameAs,n["SP_(Socialistische_Partij)"]))
g.add((np["SP_(Socialistische_Partij)"],RDF.type,dbo.party))
g.add((np["SP_(Socialistische_Partij)"],DCTERMS.description,Literal("Socialist Party (Netherlands)")))
g.add((np["SP_(Socialistische_Partij)"],OWL.sameAs,dbp['Socialist_Party_(Netherlands)']))
g.add((np["SP_(Socialistische_Partij)"],OWL.sameAs,n.SP))

#LOG
g.add((np.LOG,RDF.type,dbo.party))
#GL
g.add((np.GL,RDF.type,dbo.party))
g.add((np.GL,DCTERMS.description,Literal("GroenLinks")))
g.add((np.GL,OWL.sameAs,dbp['GroenLinks']))
#LCG
g.add((np.LCG,RDF.type,dbo.party))
#LPG
g.add((np.LPG,RDF.type,dbo.party))
#CP86
g.add((np.CP86,RDF.type,dbo.party))
g.add((np.CP86,DCTERMS.description,Literal("Centre Party '86")))
g.add((np.CP86,OWL.sameAs,dbp['Centre_Party_%2786']))
#CU
g.add((np.CU,RDF.type,dbo.party))
g.add((np.CU,DCTERMS.description,Literal("Christian Union (Netherlands)")))
g.add((np.CU,OWL.sameAs,dbp['Christian_Union_(Netherlands)']))

                            
for path in glob.glob("*.csv"):
#     print path
    paths=path.split("-")
#     print paths
    year=paths[2].replace(".csv","")
    check=paths[1]
    with open(path) as csvfile:
        if check=="2":
            csvreader = csv.DictReader(csvfile,delimiter=';')
            headers = csvreader.next()
            for row in csvreader:
                party='Partij'
                temp=row[party].replace(" ","_").replace("/","&").replace("-","&")
                locals()['row[party]']=URIRef("http://gemeenteraad.org/party/"+temp)
                for r in row:
                    if row[r]!="":
                        if r=='# Stemmen':
                            g.add((locals()['row[party]'],n['has_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        if r=='Percentage':
                            g.add((locals()['row[party]'],n['has_vote_percent'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        if r=='# Zetels':
                            g.add((locals()['row[party]'],n['has_seats'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
        if check=="1":
            csvreader = csv.DictReader(csvfile,delimiter=';')
            headers = csvreader.next()
            for row in csvreader:
                name='Naam'
                temp=row[name].replace(" ","_")
                locals()['row[name]']=URIRef("http://www.gemeentegeschiedenis.nl/gemeentenaam/"+temp)
                g.add((locals()['row[name]'],RDF.type,OWL.municipality))
                g.add((locals()['row[name]'],RDFS.label, Literal(temp)))
                for r in row:
                    if row[r]!="":
                        if r=='Amsterdamse code':
                            locals()['row[r]']=URIRef("http://www.gemeentegeschiedenis.nl/amco/"+row[r])
                            g.add((locals()['row[name]'], dbo.amsterdamCode, locals()['row[r]']))
                            g.add((locals()['row[r]'], RDFS.label, Literal(row[r],datatype=XSD.integer)))
                        elif r=='CBS code':
                            locals()['row[r]']=URIRef("http://www.gemeentegeschiedenis.nl/cbscode/0"+row[r])
                            g.add((locals()['row[r]'],RDFS.label,Literal(row['CBS code'],datatype=XSD.integer)))
                            g.add((locals()['row[name]'],n.CBScode,locals()['row[r]']))
                        elif r=='# geldige stemmen':
                            if (n['valid_vote'+'('+year+')'],DCTERMS.description,Literal("Number of valid votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['valid_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['valid_vote'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['valid_vote'+'('+year+')'],DCTERMS.description,Literal("Number of valid votes in "+year)))
                                g.add((locals()['row[name]'],n['valid_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='# ongeldige stemmen':
                            if (n['invalid_vote'+'('+year+')'],DCTERMS.description,Literal("Number of invalid votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['invalid_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['invalid_vote'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['invalid_vote'+'('+year+')'],DCTERMS.description,Literal("Number of invalid votes in "+year)))
                                g.add((locals()['row[name]'],n['invalid_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='# ongeldige':
                            if (n['invalid_vote'+'('+year+')'],DCTERMS.description,Literal("Number of invalid votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['invalid_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['invalid_vote'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['invalid_vote'+'('+year+')'],DCTERMS.description,Literal("Number of invalid votes in "+year)))
                                g.add((locals()['row[name]'],n['invalid_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='# blanco stemmen':
                            if (n['blank_vote'+'('+year+')'],DCTERMS.description,Literal("Number of blank votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['blank_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['blank_vote'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['blank_vote'+'('+year+')'],DCTERMS.description,Literal("Number of blank votes in "+year)))
                                g.add((locals()['row[name]'],n['blank_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='# ongeldige/blanco stemmen':
                            if (n['invalid_vote'+'('+year+')'],DCTERMS.description,Literal("Number of invalid votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['invalid_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['invalid_vote'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['invalid_vote'+'('+year+')'],DCTERMS.description,Literal("Number of invalid votes in "+year)))
                            if (n['blank_vote'+'('+year+')'],DCTERMS.description,Literal("Number of blank votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['blank_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['blank_vote'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['blank_vote'+'('+year+')'],DCTERMS.description,Literal("Number of blank votes in "+year)))
                                g.add((locals()['row[name]'],n['blank_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='Percentage ongeldige/blanco stemmen':
                            if (n['percent_invalid_blank'+'('+year+')'],DCTERMS.description,Literal("Percentage of invalid and blank votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['percent_invalid_blank'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['percent_invalid_blank'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['percent_invalid_blank'+'('+year+')'],DCTERMS.description,Literal("Percentage of invalid and blank votes in "+year)))
                                g.add((locals()['row[name]'],n['percent_invalid_blank'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='Percentage ongeldige stemmen':
                            if (n['percent_invalid'+'('+year+')'],DCTERMS.description,Literal("Percentage of invalid votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['percent_invalid'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['percent_invalid'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['percent_invalid'+'('+year+')'],DCTERMS.description,Literal("Percentage of invalid votes in "+year)))
                                g.add((locals()['row[name]'],n['percent_invalid'+'('+year+')'],Literal(row[r])))
                        elif r=='Percentage blanco stemmen':
                            if (n['percent_blank'+'('+year+')'],DCTERMS.description,Literal("Percentage of blank votes in "+year)) in g:
                                g.add((locals()['row[name]'],n['percent_blank'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['percent_blank'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['percent_blank'+'('+year+')'],DCTERMS.description,Literal("Percentage of blank votes in "+year)))
                                g.add((locals()['row[name]'],n['percent_blank'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='# kiesgerechtigden':
                            if (n['voters_number'+'('+year+')'],DCTERMS.description,Literal("Number of voters in "+year)) in g:
                                g.add((locals()['row[name]'],n['voters_number'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['voters_number'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['voters_number'+'('+year+')'],DCTERMS.description,Literal("Number of voters in "+year)))
                                g.add((locals()['row[name]'],n['voters_number'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='Opkomst':
                            if (n['vote_number'+'('+year+')'],DCTERMS.description,Literal("Number of vote in "+year)) in g:
                                g.add((locals()['row[name]'],n['vote_number'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['vote_number'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['vote_number'+'('+year+')'],DCTERMS.description,Literal("Number of vote in "+year)))
                                g.add((locals()['row[name]'],n['vote_number'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r=='Opkomstpercentage':
                            if (n['percent_vote'+'('+year+')'],DCTERMS.description,Literal("Percentage of vote in "+year)) in g:
                                g.add((locals()['row[name]'],n['percent_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                            else:
                                g.add((n['percent_vote'+'('+year+')'],RDF.type,RDF.Property))
                                g.add((n['percent_vote'+'('+year+')'],DCTERMS.description,Literal("Percentage of vote in "+year)))
                                g.add((locals()['row[name]'],n['percent_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
                        elif r!='Naam':
                            g.add((locals()['row[name]'],n[r.replace(" ","_").replace("/","&").replace("-","&")+'_have_vote'+'('+year+')'],Literal(row[r],datatype=XSD.integer)))
g.serialize(destination='output.ttl', format='turtle')
```
