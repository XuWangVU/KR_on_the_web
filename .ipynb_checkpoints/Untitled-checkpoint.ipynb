```python
from SPARQLWrapper import SPARQLWrapper, JSON


def similarity_ontologies(own, other):
    #also possible: list(set().union(a,b,c)), difference(), intersection()
    eq_count = 0
    dif_count = 0
    
    count_instances_other = 0
    count_instances_own = 0
    
    for result_other in other["results"]["bindings"]:
        count_instances_other +=1
    
    #check instances occurring in both and just in own
    for result_own in own["results"]["bindings"]:
        found = False
        count_instances_own += 1
        for result_other in other["results"]["bindings"]:
#             if count_instances_own == 1: count_instances_other += 1
            if result_own["obj"]["value"] == result_other["obj"]["value"]:
                eq_count += 1
                #print(result_own, result_other)
                found = True
                break
        if found == False: dif_count += 1
    
    print(count_instances_own, count_instances_other)
    print(eq_count, dif_count)
    
    #Jaccard score
    score = float(eq_count) / float(dif_count + count_instances_other)
    return score


#RELATION Municipality matching

#queries own database
sparql_own = SPARQLWrapper("https://api.druid.datalegend.net/datasets/Gemeenteraad/Gemmenteraad/services/OwnDataset/sparql")
sparql_own.setQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX owl: <http://www.w3.org/2002/07/owl#>
    
    SELECT distinct ?obj WHERE {
      ?sub a owl:municipality .
      ?sub rdfs:label ?obj
    }
""")

sparql_own.setReturnFormat(JSON)
results_own = sparql_own.query().convert()

#DBPedia queries
sparql_dbpedia = SPARQLWrapper("http://dbpedia.org/sparql")
sparql_dbpedia.setQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX dbo: <http://dbpedia.org/ontology/>
    
    SELECT distinct ?obj WHERE {
      ?sub a dbo:Municipality .
      ?sub rdfs:label ?obj
    }
""")
sparql_dbpedia.setReturnFormat(JSON)
results_dbpedia = sparql_dbpedia.query().convert()


#(Own +) LOD-a-lot queries
sparql_gem = SPARQLWrapper("https://api.druid.datalegend.net/datasets/Gemeenteraad/Gemmenteraad/services/CombineDataset/sparql")
sparql_gem.setQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX dbo: <http://dbpedia.org/ontology/>
    
    SELECT distinct ?obj WHERE {
      ?sub a dbo:Municipality .
      ?sub rdfs:label ?obj
    }
""")

sparql_gem.setReturnFormat(JSON)
results_gem = sparql_gem.query().convert()


#Score
print("\nMunicipality concept")
score = similarity_ontologies(results_own, results_dbpedia)
print("dbpedia")
print(score, equals, differents)

score = similarity_ontologies(results_own, results_gem)
print("lod-a-lot")
print(score, equals, differents)



#CBScode matching

#queries own database
sparql_own = SPARQLWrapper("https://api.druid.datalegend.net/datasets/Gemeenteraad/Gemmenteraad/services/OwnDataset/sparql")
sparql_own.setQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

    SELECT ?obj WHERE {
      ?sub <http://gemeenteraad.org/vocabulary/CBScode> ?o .
      ?o rdfs:label ?obj
    } 
""")

sparql_own.setReturnFormat(JSON)
results_own = sparql_own.query().convert()

#DBPedia queries
sparql_dbpedia = SPARQLWrapper("http://dbpedia.org/sparql")
sparql_dbpedia.setQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX dbo: <http://dbpedia.org/ontology/>
    PREFIX dbr: <http://dbpedia.org/resource/>
    
    SELECT ?obj WHERE {
      ?s dbo:municipalityCode ?obj .
      ?s dbo:country dbr:Netherlands

    }
""")
sparql_dbpedia.setReturnFormat(JSON)
results_dbpedia = sparql_dbpedia.query().convert()


#(Own +) LOD-a-lot queries
sparql_gem = SPARQLWrapper("https://api.druid.datalegend.net/datasets/Gemeenteraad/Gemmenteraad/services/CombineDataset/sparql")
sparql_gem.setQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX dbo: <http://dbpedia.org/ontology/>

    SELECT ?obj WHERE {
      ?sub dbo:municipalityCode ?obj .
    }
""")

sparql_gem.setReturnFormat(JSON)
results_gem = sparql_gem.query().convert()


#Score
print("\nCBS relation match")
score = similarity_ontologies(results_own, results_dbpedia) #NOT AVAILABLE: municipalityCode and Netherlands
print("dbpedia")
print(score, equals, differents)

score = similarity_ontologies(results_own, results_gem)
print("\nlod-a-lot")
print(score, equals, differents)
```
